## Major Project Management Models

### 1. **Waterfall Model**
- **Approach:** Linear and sequential (requirements ‚Üí design ‚Üí implementation ‚Üí testing ‚Üí deployment).
- **Best for:** Projects with fixed requirements and little expected change (e.g., construction, hardware).
- **Pros:** Clear structure, easy to manage progress.
- **Cons:** Inflexible, late discovery of issues, poor fit for evolving projects.  

---

### 2. **Agile Model**
- **Approach:** Iterative and incremental development, focusing on adaptability and customer collaboration.
- **Best for:** Software and digital projects where requirements evolve.
- **Pros:** Flexibility, faster feedback loops, customer involvement.
- **Cons:** Requires strong communication and discipline; scope creep risk.  

---

### 3. **Scrum Framework (within Agile)**
- **Approach:** Work is divided into **time-boxed sprints** (usually 2‚Äì4 weeks).
- **Roles:** Product Owner, Scrum Master, Development Team.
- **Artifacts:** Product Backlog, Sprint Backlog, Increment.
- **Best for:** Teams needing structure and accountability in iterative work.
- **Pros:** Clear roles, predictable delivery cycles, strong team collaboration.
- **Cons:** Can feel rigid; requires commitment to ceremonies (daily stand-ups, sprint reviews).  

---

### 4. **Kanban Framework (within Agile)**
- **Approach:** Continuous flow of tasks visualized on a **Kanban board** (columns like ‚ÄúTo Do ‚Üí In Progress ‚Üí Done‚Äù).
- **Best for:** Teams needing flexibility and ongoing delivery (support, operations, design).
- **Pros:** Easy to adopt, highly visual, promotes continuous improvement.
- **Cons:** Less structured than Scrum; progress can stall without discipline.  

---

## Comparison Table

| Model       | Structure         | Flexibility | Best Use Case | Key Tools |
|-------------|------------------|-------------|---------------|-----------|
| **Waterfall** | Linear phases     | Low         | Fixed-scope projects | Gantt charts |
| **Agile**    | Iterative cycles  | High        | Software, evolving needs | User stories |
| **Scrum**    | Time-boxed sprints | Medium-High | Product development | Backlogs, sprint boards |
| **Kanban**   | Continuous flow   | High        | Support/maintenance | Kanban board |

---

## Key Considerations
- **Waterfall** works well when requirements are stable but struggles with change.
- **Agile** is a philosophy; Scrum and Kanban are frameworks under Agile.
- **Scrum** suits teams that thrive on structure and accountability.
- **Kanban** suits teams that prefer flexibility and visual workflow tracking.

## What is the difference b/n timeout and assertion timeout.

### 1. **Timeout**
- A **timeout** is the maximum time allowed for a test, script, or operation to complete.  
- If the test doesn‚Äôt finish within this period, it fails due to exceeding the time limit.  
- Example: In Playwright or other test frameworks, you might set a **timeout** of 30 seconds for a test case. If the test doesn‚Äôt finish in that time, it‚Äôs marked as failed.

### 2. **Assertion Timeout**
- An **assertion timeout** is specifically about how long the framework waits for a condition (assertion) to become true.  
- Assertions often involve waiting for UI changes, elements to appear, or values to update.  
- Example: If you assert that a button should become visible, the framework will keep checking until the **assertion timeout** expires. If the condition isn‚Äôt met within that time, the assertion fails.

---

### Key Difference
| Aspect | Timeout | Assertion Timeout |
|--------|----------|-------------------|
| **Scope** | Applies to the entire test or operation | Applies only to a specific assertion/expectation |
| **Purpose** | Prevents tests from running indefinitely | Allows dynamic waiting for conditions to be met |
| **Failure Cause** | Test took too long overall | Condition wasn‚Äôt satisfied in time |

---

Think of it like this:  
- **Timeout** = "The whole exam must finish in 3 hours."  
- **Assertion Timeout** = "For this question, I‚Äôll give you 30 seconds to find the right answer."

# How **"Timeout"** and **"Assertion timeout"** interact:

### Relationship Between Timeout and Assertion Timeout
- **Timeout (test-level)**  
  - This is the *overall maximum duration* for the entire test case.  
  - Once this limit is reached, the test fails regardless of what‚Äôs happening inside (including assertions).  

- **Assertion Timeout (expect-level)**  
  - This is the *maximum wait time for a specific condition* to be satisfied.  
  - It applies only to that assertion, not the whole test.  

---

### Can Assertion Timeout Be Larger Than Timeout?
- **No practical effect**: If you set an assertion timeout longer than the test timeout, the test will still fail once the overall test timeout is reached.  
- Example:  
  - Test timeout = 10s  
  - Assertion timeout = 30s  
  - The assertion will never get its full 30s because the test itself will stop at 10s.

---

### Can Timeout Be Larger Than Assertion Timeout?
- **Yes, and this is common.**  
- Example:  
  - Test timeout = 60s  
  - Assertion timeout = 5s  
  - The test can run for up to 60s, but each assertion only waits 5s for its condition.  
- This ensures that individual checks don‚Äôt hang too long, while the test as a whole has enough time to complete.

---

### Best Practice
- **Test timeout** should usually be **greater** than assertion timeout.  
- Assertion timeout is meant for fine-grained control (like waiting for UI changes), while test timeout ensures the whole test doesn‚Äôt run indefinitely.  


Think of it like this:  
- **Timeout** = "The exam hall closes in 2 hours."  
- **Assertion timeout** = "For each question, you only get 5 minutes."  

# How **"Global timeout"**, **"Timeout"**, and **"Assertion timeout"** compare:

---

### 1. **Global Timeout**
- This is the **highest-level limit** that applies to the entire test run (all tests in a suite).  
- It ensures that your whole test execution doesn‚Äôt run indefinitely.  
- Example: If you set a **global timeout** of 30 minutes, the test runner will stop all tests once that time is exceeded, even if individual tests or assertions still have time left.

under export default {

  globalTimeout: 60*60*100,

}
---

### 2. **Timeout (Test Timeout)**
- This applies to a **single test case**.  
- It defines how long that specific test is allowed to run before failing.  
- Example: A test timeout of 60 seconds means that test must finish within 60 seconds, regardless of how many assertions it contains.

---

### 3. **Assertion Timeout**
- This applies to **individual assertions/expectations** inside a test.  
- It defines how long the framework waits for a condition to be met (like an element becoming visible).  
- Example: An assertion timeout of 5 seconds means the framework will keep checking for that condition for up to 5 seconds before failing.

---

### Comparison Table

| Type              | Scope                  | Purpose | Example |
|-------------------|------------------------|---------|---------|
| **Global Timeout** | Entire test suite run   | Prevents the whole run from hanging | Stop all tests after 30 min |
| **Timeout**        | Single test case        | Prevents one test from running too long | Fail test after 60s |
| **Assertion Timeout** | Individual assertion   | Waits for condition to be satisfied | Wait 5s for button to appear |

---

### How They Interact
- **Global timeout > Test timeout > Assertion timeout** is the usual hierarchy.  
- If **assertion timeout > test timeout**, the test will still fail once the test timeout is reached.  
- If **test timeout > global timeout**, the global timeout wins, stopping everything.  

 Think of it like this:  
- **Global timeout** = "The exam hall closes in 3 hours."  
- **Test timeout** = "Each student has 1 hour to finish their paper."  
- **Assertion timeout** = "For this question, you get 5 minutes."

# What is meant by "FLAKY"

Flaky means a testcase which fails first and then passes for the second run, such kind of test case status is called flaky.

In Playwright, **‚Äúflaky‚Äù** refers to tests that **sometimes pass and sometimes fail under the same conditions**, without any changes in the code.  

### Why tests become flaky
- **Timing issues**: The test runs before the page or element is fully ready.  
- **Unstable selectors**: Using locators that change or aren‚Äôt reliable.  
- **Environment differences**: Browser, OS, or network variations.  
- **Async gaps**: Not waiting properly for events, animations, or API responses.  

### Why it‚Äôs a problem
- Slows down CI/CD pipelines.  
- Wastes time rerunning tests instead of fixing real issues.  
- Reduces trust in the test suite.  
- Can hide legitimate bugs because teams start ignoring failures.  

### How Playwright helps
Playwright has **test retries**: you can configure it to automatically rerun a failing test. This is useful when a test is flaky and fails intermittently, though the better solution is to fix the root cause (selectors, waits, environment setup).

# Synchronous vs Asynchronous in playwright

Synchronous (Blocking)
- Tasks run one after another, in order.
- Each task must finish before the next one starts.
- If one task takes a long time (like waiting for a web page to load), everything else is stuck until it‚Äôs done.

Asynchroous (Non-Blocking)
- Tasks can start and then "wait" in the background while other tasks continue.\
- Useful for operations that tak time (network requests, file reads, timers).
- Instead of freezing the program, Javascript uses promises and the "async/await" keywords to handle these delays.

Why Playwright uses Asynchronous code?
web actions like:
- Navigate to a page (page.goto)
- Clicking a button (page.click)
- Waiting for an element (expect(locator).toBeVisible())

They all take time because, they depend on the browser and network.

If playwright were synchronous, your test would freeze until each action finished, making it inefficient and hard to manage.

Instead, playwright returns **Promises** for these actions.

¬∑ Quick Analogy
Think of synchronous like standing in a queue at a bank:

. You can't move forward until the person ahead finishes.

Asynchronous is like ordering food at a restaurant:

. You place your order, then do other things (chat, check your phone) while the
kitchen prepares it.

. When the food is ready, the waiter brings it - that's the promise being resolved.

In short:

. Synchronous = one step at a time, blocking.
. Asynchronous = tasks can wait in the background, freeing you to do other
things.

. Playwright uses async/await so your test waits for browser actions to finish
properly, without freezing everything else.

# What are workers?
workers: process.env.CI ? 1 : undefined,

What It Means
- process.env.CI ‚Üí Checks if you‚Äôre running in a Continuous Integration (CI) environment.
- Ternary operator (? :) ‚Üí
- If CI is truthy ‚Üí workers = 1 (tests run sequentially, no parallelism).
- If CI is falsy ‚Üí workers = undefined (Playwright uses its default behavior).
üëâ In CI, limiting workers to 1 avoids flaky tests caused by parallel execution (common in shared or resource-constrained environments). Locally, leaving it undefined lets Playwright decide based on your machine‚Äôs CPU cores.

Maximum Number of Workers
- Playwright does not impose a hard-coded maximum.
- By default, it uses 50% of available logical CPU cores.
- Example: On a machine with 16 logical cores, Playwright will use 8 workers unless you override it.
- You can set any integer value for workers in your config:
workers: 4,   // run with 4 parallel workers
workers: 16,  // run with 16 parallel workers (if your CPU supports it)
- The practical limit is your system‚Äôs CPU and memory capacity. Setting too many workers can cause resource contention and actually slow down tests.

Best Practice
- CI pipelines ‚Üí workers: 1 (stable, avoids flakiness).
- Local runs ‚Üí leave undefined or set to a fraction of cores (e.g., 4‚Äì8) for faster feedback.
- Heavy test suites ‚Üí experiment with higher values, but monitor CPU usage and test stability.

So:
- The 1 after ? is chosen to keep CI stable.
- The undefined after : lets Playwright auto-scale locally.
- You can set workers to any integer, but the sweet spot is usually ‚â§ number of logical cores.


# What is Fully parallel: True/False

Whatever the workers count is, but when we give the fullyparallel as false, it will run the single worker, one after the other.

if it is true, then it will run based on the workers count parallelly.

### `fullyParallel: false` (default)
- Tests in the same file run **sequentially**, even if you have multiple workers.  
- This ensures that tests in one file don‚Äôt interfere with each other (e.g., sharing state, using the same page/browser context).  
- Parallelism only happens **across different test files**.

### `fullyParallel: true`
- Tests inside a single file can run **in parallel across multiple workers**.  
- This speeds up execution if your tests are independent and don‚Äôt share resources.  
- However, if tests in the same file rely on shared setup/teardown or mutate common state, enabling this can cause flakiness.

---

### Example
```ts
// playwright.config.ts
export default defineConfig({
  fullyParallel: true,   // all tests in all files can run in parallel
});
```

- With `true`: If you have 10 tests in one file and 4 workers, Playwright will distribute those tests across workers and run them simultaneously.  
- With `false`: Those 10 tests will run one after another in a single worker, but other files can still run in parallel.

---

**Key Difference**  
- `false` ‚Üí safer, avoids conflicts, slower.  
- `true` ‚Üí faster, but requires tests to be completely independent.

**Debugger Mode: Test Runner**
1. When you want top run in debug mode, click on the "breakpoint" [a red dot on the left side], then right click on the |> icon, click on debug test & observe.

2. It starts execution and execute upto the breakpoint.

3. If you want to continue to run, we have to click on "step over / F10".

### Debugger Controls Explained
- **Continue**  
  - Resumes program execution until the next breakpoint or the end of the program.  
  - Use this when you‚Äôre done inspecting the current pause and want the program to run normally.

- **Step Over**  
  - Executes the current line of code but *does not go inside* any function calls.  
  - Useful when you trust the function being called and only care about the result, not its internal logic.

- **Step Into**  
  - Executes the current line and *enters into* any function call on that line.  
  - Use this when you want to debug inside the function being called.

- **Step Out**  
  - Runs the rest of the current function and pauses once it returns to the caller.  
  - Handy if you‚Äôve stepped into a function but realize you don‚Äôt need to see all its details.

- **Restart**  
  - Stops the current debugging session and starts the program again from the beginning.  
  - Useful if you want to re-run the test with fresh state.

- **Stop**  
  - Ends the debugging session completely.  
  - Use this when you‚Äôre finished debugging.

### Example in Playwright Context
Suppose you‚Äôre debugging a login test:
- You hit a breakpoint at `await page.fill('#username', 'admin')`.
- If you **Step Over**, it will fill the field and move to the next line.  
- If you **Step Into**, it will dive into Playwright‚Äôs internal `fill` method (usually not needed unless you‚Äôre debugging Playwright itself).  
- If you **Step Out**, it will finish the current helper function and return to the caller.  
- **Continue** will just run until the next breakpoint, maybe at `await page.click('#login')`.

Think of it like controlling how deep you want to ‚Äúpeek‚Äù into the code execution.

# Diff b/n Xpath and CSS, which is the best method to use.

1. In playwright default one is CSS Selector.

2. Perfomance wise -

- CSS is faster and natively supported by browsers.

- XPath is slower and extra parsing.

3. Readability

- CSS Simple and concise
[.products.price]

- XPah is more verbose
(//div[@class='products']//div[@class='price'])

4. Stability

- CSS is Less brittle, widely used in modern automation.

- XPath Can be brittle if DOM changes.

5. Features

- CSS Great for classes, ID's, Attributes, hierarchy.

- XPath Powerful for complex quries (e.g., Parent / Ancestor traversal)

6. Best use case

- CSS most day-to-day locators in playwright.

- XPath rare cases: when you need to traverse upward in DOM or handle tricky structures.

### To be simple:

Prefer CSS Selectors in playwright - they are faster, cleaner, and align with playwrights design philosophy.

Use XPath only when necessary - e.g., if you need to select based on parent/ancestor relationships that CSS can't handle.

Even better: Playwright recommends built-in locators like getByRole, getByText, or getByTestid for maximum stability and readability.

These are more resilient that raw CSS or XPath.

# What is a DOM Tree?

DOM - Stands for "Document Object Model"

It's a programmic representation of an HTML or XML Document.

The browser parses your HTML and builds a tree structure.

Where., the root is the <html> element, each node is an element (<div>, <span>, <p>), attribute or text.

Nodes are connected like a "Family Tree".
(Parent-> Child -> Sibling)

<html>
  <body>
    <div class="product">
      <span class="price">$29.99</span>
    </div>
  </body>
</html>

      ‚îî‚îÄ‚îÄ div (class="product")
           ‚îî‚îÄ‚îÄ span (class="price") ‚Üí "$29.99"

### Why is it called DOM?

Document: Because it represents the HTML Document.

Object: Because each element is treated as an object with properties/methods.

Model: Because it's a structured model of the document, not just raw text.

" So DOM is a structured object model of the document. "

# Variable Keywords in JavaScript

### 1. **var**
- **Introduced:** Before ES6 (2015).
- **Scope:** Function‚Äëscoped or globally scoped.
- **Hoisting:** Hoisted and initialized with `undefined`.
- **Redeclaration:** Allowed.
- **Use Case:** Legacy code; generally avoided in modern JS.

---

### 2. **let**
- **Introduced:** ES6 (2015).
- **Scope:** Block‚Äëscoped (inside `{}`).
- **Hoisting:** Hoisted but not initialized (Temporal Dead Zone).
- **Redeclaration:** Not allowed in the same scope.
- **Use Case:** For variables that need reassignment.

---

### 3. **const**
- **Introduced:** ES6 (2015).
- **Scope:** Block‚Äëscoped.
- **Hoisting:** Hoisted but not initialized (Temporal Dead Zone).
- **Redeclaration:** Not allowed.
- **Reassignment:** Not allowed (must be initialized at declaration).
- **Use Case:** For constants or references that shouldn‚Äôt be reassigned.

---

### 4. **Without Keyword (Implicit Global)**
- **Definition:** Declaring a variable without `var`, `let`, or `const` automatically creates a global variable.
- **Scope:** Global (attached to `window` in browsers).
- **Use Case:** Not recommended ‚Äî can cause accidental global pollution.

---

## Quick Comparison Table

| Keyword           | Scope        | Hoisting Behavior         | Redeclaration | Reassignment | Best Practice |
|-------------------|-------------|---------------------------|---------------|--------------|---------------|
| var               | Function/global | Hoisted, initialized `undefined` | Allowed       | Allowed      | Avoid in modern JS |
| let               | Block        | Hoisted, TDZ until declared | Not allowed   | Allowed      | Use for mutable vars |
| const             | Block        | Hoisted, TDZ until declared | Not allowed   | Not allowed  | Use for immutable refs |
| (no keyword)      | Global       | Implicit global           | Allowed       | Allowed      | Avoid (unsafe) |

---
**‚ÄúJavaScript variables can be declared using `var`, `let`, `const`, or implicitly without a keyword. Modern best practice is to use `const` by default, `let` when reassignment is needed, and avoid `var` or implicit globals.‚Äù**
